# method with argument

```python
class computer():        
    
    def __init__(self): #__init__ means to initialize the variable
        print('in init')
    
    def config(self):     
        print('i5', '8gb')

com1 = computer('i5', 8) # com1 has cpu i5 & ram is 8 

com1.config() 
```

```python
class computer:        
    
    def __init__(self,cpu,ram): # cpu & ram as argument
        print('in init')
    
    def config(self):     
        print('i5', '8gb')
        print('i7', 16)

com1 = computer('i5', 8)
com2 = computer('i7', 16)

com1.config() 
com2.config()
```

```python
class computer:        
    
    def __init__(self,cpu,ram,price): # cpu & ram as argument
        print('in init')
    
    def config(self):     
        print('i5', '8gb', 20000)
        print('i7', 16, 30000)

com1 = computer('i5', 8, 20000)
com2 = computer('i7', 16, 30000)

com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 

    def config(self):     
        print('config is', self.cpu, self.ram)

com1 = computer('i5', 8)
com2 = computer('i7', 16)

com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram, price): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 
        self.price = price 
    def config(self):     
        print('config is', self.cpu, self.ram, self.price )

com1 = computer('i5', 8, 20000)
com2 = computer('i7', 16, 30000)

com1.config() 
com2.config()
```
